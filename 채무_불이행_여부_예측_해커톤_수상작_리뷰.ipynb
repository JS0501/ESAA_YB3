##### 1. 주제 및 데이터

**주제**

다양한 금융 데이터를 활용하여 채무 불이행 가능성을 예측하는 AI 알고리즘을 개발




**데이터**

- train.csv: UID(샘플별 고유 ID), 채무와 관련된 개인 신용 데이터, 채무 불이행 여부(0-정상 상환, 1-채무 불이행)

- test.csv: UID(샘플별 고유 ID), 채무와 관련된 개인 신용 데이터

- sample_submission.csv: ID(샘플별 고유 ID), 채무 불이행 확률(예측한 채무 불이행 확률)

##### 3. 차별점 및 배울점
##### **차별점**
- 왜도 극복을 위해 제곱근 변환과 Box-Cox 사용
- Box-Cox: 비선형 변환 방법으로, 데이터의 **왜도(비대칭성)**를 줄이고 정규분포에 가깝게 만드는 방법
- 공식에 λ가 포함되는데, λ를 적절히 선택하면 꼬리가 긴 분포를 압축하거나 늘려서 좌우 대칭성을 높일 수 있음
- 이때 Box-Cox는 데이터가 양수여야만 적용 가능
- 0이나 음수 값이 있으면 변환 불가 -> Yeo-Johnson 변환 사용

##### **배울점**
- 불균형한 데이터일때 로그 변환 또는 oversampling만 시도했었는데, 제곱근 변환과 같은 새로운 방법을 시도하면 좋을 것 같음
